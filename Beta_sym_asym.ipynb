```python
import numpy as np
from scipy.stats import skewnorm
import matplotlib.pyplot as plt
samples_sym = np.random.normal(loc=0.0, scale=1.0, size=100000);
samples_asym = skewnorm.rvs(a=5, loc=0, scale=1, size=100000);
```

```python
plt.hist(samples_sym, bins=100, density=True, alpha=0.5, label='Normal')
plt.hist(samples_asym, bins=100, density=True, alpha=0.5, label='Skew Normal (a=5)')
plt.legend();
```

```python
# Calculating beta for symmetric case

# Beta via variance (symmetric case)
plus_var_sym = np.var(samples_sym[samples_sym>0])
minus_var_sym = np.var(samples_sym[samples_sym<0])
print(plus_var_sym - minus_var_sym   )

# Beta via Eq 16 via Kozlowski et al. 2016 (symmetric case)
plus_var_sym = np.sqrt(np.mean(np.abs(samples_sym[samples_sym>0]))**2*(np.pi/2))  
minus_var_sym = np.sqrt(np.mean(np.abs(samples_sym[samples_sym<0]))**2*(np.pi/2))  
print(plus_var_sym - minus_var_sym)
```

```python
# Calculating beta for asymmetric case

# Beta via variance (asymmetric case)
plus_var_asym = np.var(np.abs(samples_asym[samples_asym>0]))
minus_var_asym = np.var(np.abs(samples_asym[samples_asym<0]))
print(plus_var_asym - minus_var_asym)

# Beta via Eq 16 via Kozlowski et al. 2016 (asymmetric case)
plus_var_asym = np.sqrt(np.mean(np.abs(samples_asym[samples_asym>0]))**2*(np.pi/2))  
minus_var_asym = np.sqrt(np.mean(np.abs(samples_asym[samples_asym<0]))**2*(np.pi/2))  
print(plus_var_asym - minus_var_asym)
```
